```python
import re 
def _split_fields_tuple(t: str):
    """
    Split a single ( ... ) tuple into fields, ignoring commas inside single quotes.
    """
    fields = []
    buf = []
    in_quote = False

    for ch in t:
        if ch == "'" and (not buf or buf[-1] != "\\"):
            in_quote = not in_quote
            buf.append(ch)
        elif ch == "," and not in_quote:
            fields.append("".join(buf).strip())
            buf = []
        else:
            buf.append(ch)

    if buf:
        fields.append("".join(buf).strip())

    return fields

def parse_order_items_from_sql(sql: str):
    # Get only the VALUES block
    m = re.search(r"VALUES\s*(.*);", sql, flags=re.S | re.I)
    if not m:
        raise ValueError("Could not find VALUES block in SQL")
    values_block = m.group(1)

    # Strip comment lines
    cleaned = "\n".join(
        line for line in values_block.splitlines()
        if not line.strip().startswith("--")
    )

    # All (...) groups
    tuples = re.findall(r"\((.*?)\)", cleaned, flags=re.S)

    rows = []
    for t in tuples:
        parts = _split_fields_tuple(t)

        if len(parts) != 11:
            # skip weird/malformed rows
            continue

        # Map fields
        order_id      = int(parts[0])
        item_id       = int(parts[1])
        qty_requested = int(parts[2])

        # tag: NULL or quoted string like '{"shopee"}'
        raw_tag = parts[3].strip()
        tag = "" if raw_tag.upper() == "NULL" else raw_tag.strip("'")

        # delivery_date: 'YYYY-MM-DD'
        delivery_date = parts[4].strip().strip("'")

        # delivery_time: NULL or 'HH:MM:SS'
        raw_time = parts[5].strip()
        delivery_time = "" if raw_time.upper() == "NULL" else raw_time.strip("'")

        # team_assigned: NULL or 'Team A'
        raw_team = parts[6].strip()
        team_assigned = "" if raw_team.upper() == "NULL" else raw_team.strip("'")

        # delivered: true/false
        raw_delivered = parts[7].strip().lower()
        delivered = "TRUE" if raw_delivered == "true" else "FALSE"

        # custom: NULL or quoted text
        raw_custom = parts[8].strip()
        custom = "" if raw_custom.upper() == "NULL" else raw_custom.strip("'")

        # remarks: NULL or quoted text
        raw_remarks = parts[9].strip()
        remarks = "" if raw_remarks.upper() == "NULL" else raw_remarks.strip("'")

        # value: numeric
        value = parts[10].strip()

        rows.append([
            order_id,
            item_id,
            qty_requested,
            tag,
            delivery_date,
            delivery_time,
            team_assigned,
            delivered,
            custom,
            remarks,
            value,
        ])

    return rows

order_item_rows = parse_order_items_from_sql(SQL_ORDER_ITEMS)
order_item_rows, len(order_item_rows)
```
